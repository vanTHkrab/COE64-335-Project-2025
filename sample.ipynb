```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.DataFrame({
    'A': np.random.rand(5),
    'B': np.random.rand(5)
})
print(df)
```

```python
df.describe()
```

```python
plt.plot(df['A'], df['B'], 'o-')
plt.xlabel('A')
plt.ylabel('B')
plt.title('A vs B')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
stats.ttest_ind(df['A'], df['B'])
```
